VISUALIZATION GRAPHS

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
plt.figure(figsize=(10,5))
sns.countplot(x="Runs",data=data)
```

```python
plt.figure(figsize=(10,5))
sns.boxplot(x="Runs",data=data)
```

```python
plt.figure(figsize=(10,5))
sns.distplot(data["Runs"])
```

```python
sns.barplot(x='Mat',y="Runs",data=data)
plt.title("Runs Scored In Each match")
plt.xlabel("Matches")
plt.ylabel("Runs")
plt.show()
```

```python
sns.lineplot(x='Mat',y="Runs",data=data,marker='o',label='Runs',color='blue')
```

```python
sns.lineplot(x='Mat',y="SR",data=data,marker='o',label='Strike Rate',color='blue')
```

```python
sns.scatterplot(x='Runs',y="SR",data=data ,s=100,color='red')
plt.xlabel("Runs")
plt.ylabel("Strike Rate")
plt.title("Scatter Plot")
```
